#### Representacion Temporal y Espectral

```python
import numpy as np
import matplotlib.pyplot as plt

# Ruta del archivo generado por GNU Radio
file_path = 'Ruido_Salida_3_00.txt'

# Leer los datos del archivo
data = np.fromfile(file_path, dtype=np.float32)  # Ajusta dtype segn el formato de salida en GNU Radio

# Visualizar la seal en el dominio del tiempo
plt.figure(figsize=(10, 4))
plt.plot(data)
plt.title('Seal en el Dominio del Tiempo')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Realizar la transformada de Fourier para visualizar el espectro
fft_data = np.fft.fft(data)
fft_freq = np.fft.fftfreq(len(fft_data), 1/3.2e6)  # Ajusta la tasa de muestreo (samp_rate)

# Visualizar el espectro de la seal
plt.figure(figsize=(10, 4))
plt.plot(fft_freq, np.abs(fft_data))
plt.title('Espectro de la Seal')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Magnitud')
plt.grid(True)
plt.show()
```

#### Indice m = 3 , SNR [30, 27, 24, 21, 18, 15, 12, 9, 6, 3, 0]

```python
import numpy as np
import matplotlib.pyplot as plt

# Funcin para generar la ruta del archivo segn el ndice
def generate_file_path(prefix, suffix, index):
    return f'{prefix}{index:02d}{suffix}'

# Bucle para iterar desde 30 hasta 0
for i in range(30, -1, -3):
    file_path_message = generate_file_path('Ruido_Msj_3_', '.txt', i)
    file_path_modulated = generate_file_path('Ruido_Salida_3_', '.txt', i)
    
    try:
        # Cargar datos del archivo de mensaje original como bytes
        with open(file_path_message, 'rb') as file:
            original_data = np.fromfile(file, dtype=np.float32)

        # Cargar datos del archivo de seal modulada como bytes
        with open(file_path_modulated, 'rb') as file:
            modulated_data = np.fromfile(file, dtype=np.float32)

        # Graficar en modo XY
        plt.figure(figsize=(10, 6))
        plt.plot(original_data, modulated_data, 'o', markersize=1)
        plt.title(f'Grfica XY: Seal del Mensaje vs Seal Modulada (SNR {i} y m = 3)')
        plt.xlabel('Seal del Mensaje Original')
        plt.ylabel('Seal Modulada')
        plt.grid(True)
        
        # Guardar el grfico con el nombre deseado
        plot_filename = f'3_{i:02d}.png'
        plt.savefig(plot_filename)
        plt.close()  # Cierra la figura para liberar memoria

    except Exception as e:
        print(f"Error al cargar o procesar los archivos para index {i}: {e}")
```
